```python
def primo(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
lista = [1,2,4,7,9,11]
for i in lista:
  print(primo(i),i)
```

```python
def factorial(n):
  factorial = 1
  if n == 0 or n == 1:
    return 1
  else:
    for i in range(1,n+1):
      factorial *= i
    return factorial
```

```python
factorial(5)
```

```python
def perfecto(n):
  lista_divisores = []
  for i in range(1,n):
    if n % i == 0:
      lista_divisores.append(i)
  if sum(lista_divisores) == n:
    return True
  else:
    return False
```

```python
n = 6
perfecto(n)
```

```python
#Factorial

def factorial_recursivo(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial_recursivo(n-1)
```

```python
factorial_recursivo(4)
```

```python
#MCD

def mcd(m,n):
  if m == n:
    return n
  elif m > n:
    return mcd(n,m-n)
  elif m < n:
    return mcd(n-m , m)
```

```python
mcd(89,12)
```

```python
def serie(n):
  if n == 1:
    return 1/n
  else:
    if n%2 == 0:
      return -1/n + serie(n-1)
    else:
      return 1/n + serie(n-1)
```

```python
serie(6)
```
